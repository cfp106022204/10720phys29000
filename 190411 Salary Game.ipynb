```python
# People have no money can't play any game.
import numpy as np
import matplotlib.pyplot as plt
n,m,k,dm=500,40,10000,10
money=[m for i in range(0,n)]

for i in range(k):
    p1=p2=0
    while (p1==p2 or money[p1]==0 or money[p2]==0):
        p1=np.random.randint(0,n)
        p2=np.random.randint(0,n)
    
    win=np.random.randint(0,2)
    if (win==0 ):
        money[p1]-=dm
        money[p2]+=dm
    elif (win==1 ):
        money[p1]+=dm
        money[p2]-=dm

bins=10
plt.clf()
plt.hist(money,bins)
plt.xlabel("Money")
plt.ylabel("Number of Student")
plt.title("Salary Game")
plt.ylim(0,n)
plt.show()
```

```python
# People with no money can win money again.
import numpy as np
import matplotlib.pyplot as plt
n,m,k,dm=500,40,10000,10
money=[m for i in range(0,n)]

for i in range(k):
    p1=p2=0
    while (p1==p2):
        p1=np.random.randint(0,n)
        p2=np.random.randint(0,n)
    
    win=np.random.randint(0,2)
    if (win==0 and money[p1]!=0):
        money[p1]-=dm
        money[p2]+=dm
    elif (win==1 and money[p2]!=0):
        money[p1]+=dm
        money[p2]-=dm

bins=10
plt.clf()
plt.hist(money,bins)
plt.xlabel("Money")
plt.ylabel("Number of Student")
plt.title("Salary Game")
plt.ylim(0,n)
plt.show()
```
