```python
import numpy as np
import matplotlib.pyplot as plt
import math

num=100
h=6.62*10**-34
c=3*10**8
k=1.38*10**-23

v=np.linspace(10**0,10**15.5,num,endpoint=True)
T=5000
B=2*h*v**3/c**2/(math.e**(h*v/k/T)-1)

plt.plot(v,B)
plt.xlabel("f (Hz)")
plt.ylabel("Intensity")

dB=np.random.normal(0.,10**-8.5,num)
B+=dB
plt.plot(v,B)


plt.show()
```
