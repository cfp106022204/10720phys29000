```python
fp = open('Alice.txt','r',encoding='UTF-8')
dict = {}
line = fp.readline()

while line:
    line = fp.readline()
    s = line.split()
    for word in s:
        if word not in dict:
            dict[word] = 1
        else:
            dict[word] += 1
            
for word, count in dict.items():
    print(word+": "+str(count))
    
fp = fp.close()
```

```python
import matplotlib.pyplot as plt
import numpy as np
fp = open('Alice.txt','r',encoding='UTF-8')
dict = {}
line = fp.readline()

while line:
    line = fp.readline()
    splt = line.split()
    for word in splt:
        if word not in dict:
            dict[word] = 1
        else:
            dict[word] += 1

fp = fp.close()

num = []
for key in dict:
    num.append(dict[key])
num.sort()
num.reverse()

#x=np.linspace(1,len(num),len(num))
#plt.loglog(x,num)
plt.loglog(range(len(num)),num)
plt.title('Zipf\'s Law')
plt.xlabel('Rank of the word')
plt.ylabel('Number of the occurance')
plt.show()
```
